### find_element 와 find_elements 의 차이점


find_element 및 find_elements는 웹 페이지에서 요소를 찾기 위해 Selenium WebDriver API에서 제공하는 두 가지 방법입니다.


find_element 메소드는 페이지에서 단일 웹 요소를 찾는 데 사용됩니다. 지정된 기준과 일치하는 첫 번째 요소를 반환합니다. 예를 들어 특정 클래스 이름을 가진 단일 요소를 찾으려면 (By.CLASS_NAME, 'class_name') 인수와 함께 find_element 메서드를 사용합니다.


반면 'find_elements' 메서드는 페이지에서 여러 웹 요소를 찾는 데 사용됩니다. 지정된 기준과 일치하는 모든 요소의 목록을 반환합니다. 예를 들어 특정 클래스 이름을 가진 모든 요소를 ​​찾으려면 (By.CLASS_NAME, 'class_name') 인수와 함께 find_elements 메서드를 사용합니다.


지정된 기준과 일치하는 요소가 없으면 find_element 메서드는 NoSuchElementException을 발생시키는 반면 find_elements 메서드는 빈 목록을 반환합니다. 또한 find_element 메서드는 기준과 일치하는 첫 번째 요소만 반환하고 find_elements 메서드는 기준과 일치하는 모든 요소를 ​​반환합니다.
